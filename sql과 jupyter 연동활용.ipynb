```python
## sakila rental 테이블에서 월별 렌탈 수 출력 
## 날짜는 오름차순 정렬
## dataframe
# pandas sql 
```

```python
sql='''select date_format(rental_date, '%Y-%m') as sdate, count(rental_id) as count from rental group by sdate order by sdate;
'''
```

```python
## date_format() -> sql 함수 
```

```python
pd.read_sql(sql,sdb)
```

```python
rental_df=pd.read_sql('select* from rental', sdb)
```

```python
rental_df
```

```python
rental_df=rental_df[['rental_date','rental_id']]
```

```python
rental_df['month']=rental_df['rental_date'].apply(lambda x: str(x)[:7])
```

```python
rental_df
```

```python
rental_df.groupby('month')['rental_id'].count()
```

```python
### payment table 
# 월별 전체 매출, 평균 매출, 결제 개수
# 위의 값을 출력하고 
# 위의 각각의 값을 dataframe의 칼럼값으로 해서 출력 
# pandas sql
```

```python
payment=pd.read_sql('select * from payment',sdb)
```

```python
payment
```

```python
payment_df=payment[['amount','payment_date','payment_id']]
```

```python
payment_df['month']=payment_df['payment_date'].apply(lambda x: str(x)[:7])
```

```python
payment_df
```

```python
a=payment_df.groupby('month')['amount'].sum()
```

```python
b=payment_df.groupby('month')['amount'].mean()
```

```python
c=payment_df.groupby('month')['amount'].count()
```

```python
total=pd.concat([a,b,c], axis=1)
total
```

```python
sql='''select date_format(payment_date, '%Y-%m') as sdate, avg(amount), sum(amount), count(payment_id) from payment group by sdate;
'''
```

```python
pd.read_sql(sql, sdb)
```

```python
import numpy as np
payment_df.groupby('month').agg({'amount':[np.mean, np.sum], 'payment_id':np.size})
```
